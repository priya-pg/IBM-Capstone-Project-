```python
#importing required libraries
import pandas as pd
import requests
```

```python
#downloading the dataset

import requests

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"  
filename = "survey_data.csv"

response = requests.get(url)

# Check request success
if response.status_code == 200:
    with open(filename, "wb") as f:
        f.write(response.content)   # save CSV as binary
    print(f"File saved as {filename}")
else:
    print("Failed to download:", response.status_code)
```

```python
#reading the dataset into pandas
df = pd.read_csv(filename) 
```

```python
#Printing the mean age of survey participants 

import re
def parse_age(age_str): #----------> The Age column is object, and has ranges instead of a number, hence we pass this function
    if "Under" in age_str:
        return 17  
    else:
        # Extract numbers from the string
        nums = list(map(int, re.findall(r"\d+", age_str)))
        if len(nums) == 1:
            return nums[0]
        elif len(nums) == 2:
            return sum(nums) / 2  # midpoint
        else:
            return None

df["age_numeric"] = df["Age"].apply(parse_age)

mean_age=df["age_numeric"].mean()
print(f"The mean age of the participants in the survey is {mean_age}")
```
